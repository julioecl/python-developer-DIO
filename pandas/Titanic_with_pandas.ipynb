```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from posixpath import sep
df = pd.read_csv('/content/sample_data/Titanic-Dataset.csv')
```

```python
df.head()
```

Renomeando as colunas:

```python
df.columns = ['IDpassageiro', 'Sobrevivente', 'ClassePassagem', 'Nome', 'Genero', 'Idade' ,'NrIrmaosCojuges', 'NrPaisFilhos', 'NrTicket', 'Tarifa', 'Cabine', 'PortoEmbarque' ]
```

Agrupando os dados por gênero e sobrevivente e contando o número de passageiro pela id:

```python
df.groupby(['Genero', 'Sobrevivente'])['IDpassageiro'].count()
```

A média de idade de quem sobreviveu (1) e quem não sobreviveu (0):

```python
df.groupby('Sobrevivente')['Idade'].mean()
```

Encontrando o passageiro mais velho:

```python
max_idade = df['Idade'].max()
```

```python
mais_velho = df.loc[df['Idade'] == max_idade]
mais_velho.head()
```

Encontrando o passageiro mais jovem:

```python
min_idade = df['Idade'].min()
```

```python
mais_jovem = df.loc[df['Idade'] == min_idade]
mais_jovem.head()
```
